```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv('data/data.csv')
df.head()

# Fix dataset collumns 
df.drop('Unnamed: 32', inplace=True, axis=1)

print(df.info())
print(df.isna().sum())
```

```python
# Replace string data by numbers
df['diagnosis'] = df['diagnosis'].replace('M', 0)
df['diagnosis'] = df['diagnosis'].replace('B', 1)

df
```

```python
_df = df[[
    'diagnosis',
    'texture_mean',
    'area_mean',
    'fractal_dimension_mean',
    'concavity_mean'
    ]]

scatter_matrix(_df, diagonal='hist', figsize=(14, 14))
```

```python
import matplotlib.pyplot as plt
df.boxplot(column=['diagnosis',
    'texture_mean',
    'area_mean',
    'fractal_dimension_mean',
    'concavity_mean'])
```

```python
#Separando as variÃ¡veis entre preditoras (x) e variÃ¡vel alvo (y)
y = df['diagnosis']
x = df.drop('diagnosis', axis= 1)
```

```python
#Criando os conjuntos de dados de teste e treino
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size =0.3)
```

```python
x.shape, y.shape, x_treino.shape, y_treino.shape, x_teste.shape, y_teste.shape
```

```python
from sklearn.metrics import (recall_score,
                             accuracy_score,
                             precision_score,
                             f1_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
# instancia um DecisionTreeClassifier
modelo= DecisionTreeClassifier()

# treina o classificador
modelo.fit(x_treino, y_treino)
pred = modelo.predict(x_teste)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

sns.heatmap(confusion_matrix(y_teste, pred), cmap="gray_r", annot=True, fmt='2.0f')
plt.title('Decision Tree')
plt.ylabel('P R E V I S T O')
plt.xlabel('AMOSTRA')
plt.show()
```

```python
resultado = modelo.score(x_teste, y_teste)
#AcurÃ¡cia: 
print("ACC :%.2f" %(accuracy_score(y_teste,pred))) 

#RevocaÃ§Ã£o:
print("Recall :%.2f" %(recall_score(y_teste,pred)))  

#PrecisÃ£o:
print("Precision :%.2f" %(precision_score(y_teste,pred)))  

#F1-score:
print("F1-score :%.2f" %(f1_score(y_teste,pred)))
```
